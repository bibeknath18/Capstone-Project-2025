```python
df["Timestamp"] = pd.to_datetime(df["LastUpdatedDate"] + " " + df["LastUpdatedTime"], format="%d-%m-%Y %H:%M:%S")
df["Hour"] = df["Timestamp"].dt.hour
df["DayOfWeek"] = df["Timestamp"].dt.dayofweek
```

```python
# Feature Mappings
vehicle_weight_map = {'car': 1.0, 'bike': 0.5, 'bus': 1.5, 'truck': 1.8}
traffic_score_map = {"low": 0.3, "medium": 0.6, "high": 1.0}
df["VehicleTypeWeight"] = df["VehicleType"].map(vehicle_weight_map).fillna(1.0)
df["TrafficScore"] = df["TrafficConditionNearby"].map(traffic_score_map).fillna(0.5)
df = df.sort_values(by="Timestamp").reset_index(drop=True)
```

```python
# Model 1: Smoothed Linear Model
BASE_PRICE = 10
alpha = 2
smoothing_factor = 0.6
prices_model1 = [BASE_PRICE]
for i in range(1, len(df)):
    occ = df.loc[i, "Occupancy"]
    cap = df.loc[i, "Capacity"]
    raw_price = prices_model1[-1] + alpha * (occ / cap)
    smoothed_price = smoothing_factor * prices_model1[-1] + (1 - smoothing_factor) * raw_price
    prices_model1.append(smoothed_price)
df["Price_Model_1"] = prices_model1
```

```python
# Model 2: Demand-based dynamic pricing
alpha_d, beta, gamma, delta, epsilon, lambda_d = 1.0, 0.5, 0.7, 1.0, 0.8, 0.6

df["RawDemand"] = (
    alpha_d * (df["Occupancy"] / df["Capacity"])
    + beta * df["QueueLength"]
    - gamma * df["TrafficScore"]
    + delta * df["IsSpecialDay"]
    + epsilon * df["VehicleTypeWeight"]
    + 0.2 * (df["Hour"] / 24.0)
)
```

```python
# Normalize demand per parking lot
def normalize_per_lot(sub_df):
    if sub_df["RawDemand"].nunique() <= 1:
        sub_df["NormalizedDemand"] = 0.5  # Constant fallback
    else:
        min_d = sub_df["RawDemand"].min()
        max_d = sub_df["RawDemand"].max()
        sub_df["NormalizedDemand"] = (sub_df["RawDemand"] - min_d) / (max_d - min_d)
    return sub_df
```

```python
df = df.groupby("SystemCodeNumber").apply(normalize_per_lot).reset_index(drop=True)

df["Price_Model_2"] = (
    BASE_PRICE * (1 + lambda_d * df["NormalizedDemand"])
).clip(lower=0.5 * BASE_PRICE, upper=2.0 * BASE_PRICE)
```

```python
# Model 3: Competitive pricing
coords = df[["Latitude", "Longitude"]].values
nn = NearestNeighbors(radius=0.01, algorithm="ball_tree").fit(coords)
distances, indices = nn.radius_neighbors(coords)
```

```python
nearby_avg_prices = []
for i, neighbors in enumerate(indices):
    neighbors = [n for n in neighbors if n != i]
    if neighbors:
        avg_price = df.loc[neighbors, "Price_Model_2"].mean()
    else:
        avg_price = df.loc[i, "Price_Model_2"]
    nearby_avg_prices.append(avg_price)

df["NearbyAvgPrice"] = nearby_avg_prices
```

```python
def competitive_price(row):
    price, nearby_price = row["Price_Model_2"], row["NearbyAvgPrice"]
    occ_ratio = row["Occupancy"] / row["Capacity"]
    if occ_ratio > 0.9 and nearby_price < price:
        return price * 0.9
    elif nearby_price > price:
        return min(price * 1.1, 20)
    else:
        return price
```

```python
df["Price_Model_3"] = df.apply(competitive_price, axis=1)

# Filter a sample parking lot
sample_lot = df["SystemCodeNumber"].iloc[0]
df_lot = df[df["SystemCodeNumber"] == sample_lot].dropna(
    subset=["Price_Model_1", "Price_Model_2", "Price_Model_3"]
).reset_index(drop=True)

# Choose a better lot with variable Model 2 prices
valid_lots = df.groupby("SystemCodeNumber")["Price_Model_2"].nunique().sort_values(ascending=False)
best_lot = valid_lots[valid_lots > 1].index[0]

# Use that lot for plotting
df_lot = df[df["SystemCodeNumber"] == best_lot].dropna(
    subset=["Price_Model_1", "Price_Model_2", "Price_Model_3"]
).reset_index(drop=True)

print("Plotting for lot:", best_lot)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
output_notebook()

source = ColumnDataSource(df_lot)

p = figure(x_axis_type="datetime", title=f"Dynamic Pricing Models – Lot: {best_lot}", width=800, height=350)
p.line(x="Timestamp", y="Price_Model_1", source=source, color="blue", line_width=2, legend_label="Model 1")
p.line(x="Timestamp", y="Price_Model_2", source=source, color="green", line_width=2, legend_label="Model 2")
p.line(x="Timestamp", y="Price_Model_3", source=source, color="red", line_width=2, legend_label="Model 3")
p.xaxis.axis_label = "Time"
p.yaxis.axis_label = "Price (₹)"
p.legend.location = "top_left"
show(p)
```
